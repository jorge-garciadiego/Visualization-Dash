```python
import numpy as np
import plotly.figure_factory as ff
import plotly.offline as pyo
```

Distribution Plots, or Distplots, typically layer three plots on top to eachone.
- a Histogram, where each data point is placed inside a bin of similar values
- The second is a rug plot - marjs are placed along the x-axis for every datapoint, which lets you see the distribution of values inside each bin
- a 'Kernel density estimate' of KDE line that tries to describe the shape of the distribution

```python
x = np.random.randn(1000)
hist_data = [x]
group_labels = ['distplot']

fig = ff.create_distplot(hist_data=hist_data, group_labels=group_labels)
fig.show()
```

```python
x1 = np.random.randn(200) - 2
x2 = np.random.randn(200)
x3 = np.random.randn(200) +2
x4 = np.random.randn(200) +4

hist_data = [x1, x2, x3, x4]
group_labels = ['x1', 'x2', 'x3', 'x4']

fig = ff.create_distplot(hist_data=hist_data, group_labels=group_labels, bin_size=[.2,.1,.3,.4])
fig.show()
```

```python
'''
Lets visualize the Snodgrass and Twain example
'''

snodgrass = [.209, .205, .196, .210, .202, .207, .224, .223, .220, .201]
twain = [.225, .262, .217, .240, .230, .229, .235, .217]

hist_data = [snodgrass, twain]
group_labels = ['Snodgrass', 'Twain']

fig = ff.create_distplot(hist_data, group_labels, bin_size=[0.005, 0.005])
fig.show()
```
